```python
""" On rennome les colonnes"""
df6xlsx.rename(columns = {"Unnamed: 0" : "num_dep", "Unnamed: 1" : "nom_dep", "Unnamed: 9" : "pop_totale"}, inplace=True)
```

```python
""" On regarde la proportion de valeur manquante"""
df6xlsx.isna().mean()
```

```python
"""On remplace les colonnes de type int ou float manquantes par la médiane"""

for col in df6xlsx.select_dtypes(["int64", "float64"]).columns :
    df6xlsx[col].fillna(df6xlsx[col].median(), inplace=True)
```

```python
""" On observe qu'à partir de la ligne 106, ça n'a plus d'întérêt donc on supprime """
df6xlsx.loc[df6xlsx["num_dep"].isna() | (df6xlsx["num_dep"] == 0)]
```

```python
df6xlsx.drop(df6xlsx.index[105::], inplace=True)
df6xlsx
```
