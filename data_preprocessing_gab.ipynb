```python
""" On rennome les colonnes"""
washed_df6xlsx = df6xlsx.rename(columns = {"Unnamed: 0" : "num_dep", "Unnamed: 1" : "nom_dep", "Unnamed: 9" : "pop_totale"})
```

```python
""" On regarde la proportion de valeur manquante"""
washed_df6xlsx.isna().mean()
```

```python
"""On remplace les colonnes de type int ou float manquantes par la médiane"""

for col in washed_df6xlsx.select_dtypes(["int64", "float64"]).columns :
    washed_df6xlsx[col].fillna(washed_df6xlsx[col].median(), inplace=True)
```

```python
""" On observe qu'à partir de la ligne 106, ça n'a plus d'întérêt donc on supprime """
washed_df6xlsx.loc[washed_df6xlsx["num_dep"].isna() | (washed_df6xlsx["num_dep"] == 0)]
```

```python
""" On drop les colonnes inutiles"""

washed_df6xlsx.drop(washed_df6xlsx.index[105::], inplace=True)
washed_df6xlsx
```

```python
"""et on vire les indications inutiles"""
row_to_drop = washed_df6xlsx.loc[(washed_df6xlsx["num_dep"]=="P")|(washed_df6xlsx["num_dep"]=="M")].index

washed_df6xlsx.drop(row_to_drop, inplace=True)
washed_df6xlsx.isna().mean()
```
