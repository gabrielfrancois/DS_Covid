# Lavage de url4cvs (concernant les établissements)

```python
"""on fait en sorte d'avoir des noms pour chaque colonnes pour pouvoir travailler sur le dataFrame"""
names_columns = ["finess", "etalab"]+[str(x) for x in range(30)]

etablissement = pd.read_csv(url4csv, sep=";", header = 0, names=names_columns)
```

```python
washed_etablissement = etablissement.copy()
pd.set_option('display.max_columns', None)#pour pouvoir afficher tout le data frame et savoir quoi garder
washed_etablissement.head()
```

```python
"""On sélectionne les colonnes intéressantes """
washed_etablissement = washed_etablissement.loc[:,["11","12","17","23",]]
washed_etablissement.head()
```

```python
washed_etablissement.rename(columns={"11" : "num_dep", "12" : "nom_dep", "17" : "type_centre", "23" : "statut_centre"}, inplace=True)
washed_etablissement.head()
```

```python
"""On souhaite sélectionner les lieux pouvant avoir un lien avec la guérison/dépistage du covid uniquement"""
word_covid = ["hospi", "pharma", "maison de santé", "Laboratoire d'Analyses"]
washed_etablissement["etat_soin_covid"] = washed_etablissement["type_centre"].str.contains(pat="|".join(word_covid), case=False, na=False).astype(int)
```

```python
washed_etablissement.loc[washed_etablissement["etat_soin_covid"]==1]["type_centre"].unique()
```

```python
"""Certaines cathégories d'établissement se sont glissée dans notre large sélection, on les retire donc"""
line_to_drop = (["Centre Hospitalier Spécialisé lutte Maladies Mentales", "Maison de Santé pour Maladies Mentales", 
"Laboratoire pharmaceutique préparant délivrant allergènes", "Syndicat Inter Hospitalier (S.I.H.)"]
    )
washed_etablissement.loc[washed_etablissement["type_centre"].isin(line_to_drop),"etat_soin_covid"] = 0
```

```python
washed_etablissement.sample(5)

```

```python
""" Désormais on n'a plus qu'à joindre par département, message pour YOUNES : quand tu fera la base de donnée commune, n'oublie pas de faire une pondération par la population, pas sûr que saint Pierre et Miquelon soient les plus mall lotis avec deux établissement de soins"""
washed_etablissement = washed_etablissement.groupby(["nom_dep", "num_dep"])["etat_soin_covid"].sum().reset_index()
washed_etablissement.sort_values(by = "etat_soin_covid",ascending=False)
```
